```python
# TBDY 2019'a göre "Yatay Elastik Tasarım Spektrumu" hesabı
import numpy as np
# INPUTS
shr ="Sakarya"    # Şehir
zmn ="ZE"         # Zemin Sınıfı
Ss = 1.518 
Fs = 0.800
S1 = 0.417  
F1 = 2.366
Ti = float(input("Enter building period:"))

# CALCULATIONS
Sds = Ss*Fs
Sd1 = S1*F1
Ta = 0.2*Sd1/Sds 
Tb = Sd1/Sds
Tl = 6.0

# define a function for spectrum value calculation
def spec(Ti):
    if Ti <= Ta:
        Saei = (0.4 + 0.6*Ti/Ta)*Sds
    elif Ti > Ta and Ti <= Tb:
        Saei = Sds
    elif Ti > Tb and Ti <= Tl:
        Saei = Sd1/Ti
    else: 
        Saei = Sd1*Tl/Ti**2
        
    return Saei


# Calculate an Sae value for any given T
Sae = spec(Ti)                # [g]; elastik tasarım spektral ivmeleri
print(f"For T= {Ti} s, Sae={np.round(Sae, 2)} g")


# Calculate all Sae values for any  given T range

T = np.arange(0.01, 7.1, 0.1)

Sae = list()    # [g]; elastik tasarım spektral ivmeleri
for Ti in T:
    Saei = spec(Ti)
    Sae.append(Saei)

print("Sae=", Sae) 
```

```python
# Plot spectrum
import matplotlib.pyplot as plt

fig, grfk = plt.subplots()
grfk.plot(T,Sae, color="r")
grfk.set(xlabel='Periyot [s]', ylabel='Sae(T), [g]', title='Yatay Elastik Tasarım Spektrumu (TBDY 2019)')
grfk.set_xlim([0 ,7.5])
grfk.set_ylim([0, 1.5])
grfk.grid()
plt.show()
```

```python
# # Bunu tek grafik çizdirdikten sonra anlat. Bunları revize et
Sae = np.array([Sae]).T

fig, grfk = plt.subplots()
grfk.plot(T,Sae, color="r", label="Sae(T) [g]")
grfk.plot(T,Sae*9.81, color="b",label="Sae(T) [m/s2]")
grfk.set(xlabel='Periyot [s]', ylabel='Sae(T)', title='Yatay Elastik Tasarım Spektrumu (TBDY 2019)')
grfk.set_xlim([0, 7.5])
grfk.set_ylim([0, 1.2*max(Sae*9.81)])
grfk.grid()
grfk.legend()
plt.show()
```

```python
# alt alta
fig, grfk = plt.subplots(nrows=2, ncols=1) # Alt alta çizim, üstteki satırı sil
grfk[0].plot(T, Sae, color="r", linewidth=1)
grfk[0].set(xlabel='Periyot [s]', ylabel='Sae(T), [g]', title="Grafik 1")
grfk[0].grid() ;
grfk[1].plot(T, Sae*9.81, color="b", linewidth=1)
grfk[1].set(xlabel='Periyot [s]', ylabel='Sae(T), [m/s2]', title="Grafik 2")
grfk[1].grid()
# fig.suptitle("aaaa")
fig.tight_layout()
plt.show()
```
